```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
diabetes_X,diabetes_y = load_diabetes(return_X_y = True)
```

```python
pd.DataFrame(diabetes_X)
```

```python
pd.Series(diabetes_y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(diabetes_X),pd.Series(diabetes_y),test_size = 0.25,random_state = 100)

X_train = X_train.reset_index(drop = True)
X_test = X_test.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)

print("Size of X_train: ",X_train.shape)
print("Size of X_test: ",X_test.shape)
print("Size of y_train: ",y_train.shape)
print("Size of y_test: ",X_test.shape)


```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
# Mean squared error: (sum(y_actual-y_predicted)^2)/2
# R squared: (sum(y_actual-y_predicted)^2)/sum(y_actual-y_mean)^2)
print('Mean squared error for training set:%.2f'%mean_squared_error(y_train,y_train_pred))
print('Coefficient of determination set:%.2f'%r2_score(y_train,y_train_pred))
```

```python
y_test_pred = model.predict(X_test)
```

```python
print('Mean squared error for training set:%.2f'%mean_squared_error(y_test,y_test_pred))
print('Coefficient of determination set:%.2f'%r2_score(y_test,y_test_pred))
```

```python
print('Coefficients:',model.coef_)
```

```python
data = pd.read_csv('demo2_data.csv')
data
```

```python
data.waterfront = data.waterfront.map({'Yes' : 1,'No' : 0})
data
```

```python
# pearson correlation 
# sum((x-x_mean)*(y-y_mean))/square_root(sum(x-x_mean)^2)* sum((y-y_mean)^2)
corrs = data.corr()
plt.figure(figsize = (16,12))
sns.heatmap(corrs,annot = True,fmt = '.2f')
```

```python
X = data.iloc[:,2:] 
y = data.iloc[:,1:2]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 100)
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)

print("Size of X_train:",X_train.shape)
print("Size of X_test:",X_test.shape)
print("Size of y_train:",y_train.shape)
print("Size of y_test:",y_test.shape)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
regressor = LinearRegression()
regressor.fit(X_train_scaled,y_train)
```

```python
y_train_pred = regressor.predict(X_train_scaled)
```

```python
print('Mean squared error for training set:%.2f'%mean_squared_error(y_train,y_train_pred))
print('Coefficient of determination set:%.2f'%r2_score(y_train,y_train_pred))
```

```python
y_test_pred = regressor.predict(X_test_scaled)
```

```python
print('Mean squared error for training set:%.2f'%mean_squared_error(y_test,y_test_pred))
print('Coefficient of determination set:%.2f'%r2_score(y_test,y_test_pred))
```
